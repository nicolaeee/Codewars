```python
#AfiÈ™area data-set
import pandas as pd


#Citirea fiÈ™ierului
file_path = "/content/freelancer_earnings_bd.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

#AfiÈ™area primelor rÃ¢nduri
df.head()
```

```python
#Creare matrice din datele alese
import numpy as np

# SelectÄƒm coloanele numerice pentru analizÄƒ
selected_data = df[["Earnings_USD", "Hourly_Rate", "Job_Success_Rate", "Client_Rating"]]

# Convertim Ã®ntr-o matrice numpy
data_matrix = selected_data.to_numpy()

print("Matricea numericÄƒ (primele 5 rÃ¢nduri):")
print(data_matrix[:5])
```

```python
# Media
average_values = selected_data.mean()
print("\nðŸ“Š Media valorilor:")
print(average_values)
```

```python
# Mediana
median_values = selected_data.median()
print("\nðŸ“Š Mediana valorilor:")
print(median_values)
```

```python
# Valoarea cea mai frecventa
mode_values = selected_data.mode().iloc[0]
print("\nðŸ“Š Modul (valoarea cea mai frecventÄƒ):")
print(mode_values)
```

```python
#Range min max
range_values = selected_data.max() - selected_data.min()
print("\nðŸ“Š Range (domeniul valorilor):")
print(range_values)
```

```python
#DeviaÈ›ia standard
std_dev = selected_data.std()
print("\nðŸ“Š DeviaÈ›ia standard:")
print(std_dev)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# HistogramÄƒ
selected_data.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.suptitle("ðŸ“Š HistogramÄƒ pentru fiecare coloanÄƒ numericÄƒ")
plt.tight_layout()
plt.show()

# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=selected_data)
plt.title("ðŸ“¦ Boxplot pentru fiecare caracteristicÄƒ numericÄƒ")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# Scatter plot Ã®ntre douÄƒ coloane relevante
plt.figure(figsize=(8, 5))
sns.scatterplot(data=selected_data, x="Hourly_Rate", y="Earnings_USD")
plt.title("ðŸ”µ Scatter plot: Hourly Rate vs Earnings")
plt.grid(True)
plt.show()
```

```python
correlation_matrix = selected_data.corr()

print("\nðŸ”— CorelaÈ›ii Ã®ntre coloane:")
print(correlation_matrix)

# Reprezentare graficÄƒ a corelaÈ›iei
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("ðŸ”— Matricea corelaÈ›iilor")
plt.show()
```
